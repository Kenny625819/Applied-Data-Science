```python
# ============================================================
# MESCC Meta-analysis
# Clean vs expanded vs complete paralysis
# Google Colab full submission code
# Complete paralysis subgroup is handled descriptively
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# ============================================================
# Global style
# ============================================================

BASE_FONT = 14
TITLE_FONT = 16
FIGSIZE = (8, 8)

plt.rcParams.update({
    "font.size": BASE_FONT,
    "axes.titlesize": TITLE_FONT,
    "axes.labelsize": BASE_FONT,
    "xtick.labelsize": BASE_FONT,
    "ytick.labelsize": BASE_FONT,
    "legend.fontsize": BASE_FONT,
    "font.family": "DejaVu Sans",
    "axes.edgecolor": "black",
    "axes.linewidth": 1.2,
    "xtick.color": "black",
    "ytick.color": "black",
    "text.color": "black",
    "axes.labelcolor": "black",
})

# ============================================================
# Dataset
# event = postoperative ambulatory recovery
# ============================================================

data = pd.DataFrame({
    "Study": [
        "Kim 2011",
        "Tateiwa 2019",
        "Ohashi 2017",
        "Kinoshita 2023",
        "Li 2024",
        "Wanman 2025",
        "Kim HCC 2011",
        "Gokaslan 1998",
        "Park NSCLC 2016",
        "Younsi 2021"
    ],
    "n": [57, 31, 82, 113, 37, 111, 14, 13, 39, 28],
    "event": [39, 17, 50, 70, 22, 61, 4, 10, 23, 7],
    "analysis_group": [
        "Clean cohort",
        "Clean cohort",
        "Clean cohort",
        "Clean cohort",
        "Clean cohort",
        "Clean cohort",
        "Expanded cohort",
        "Expanded cohort",
        "Expanded cohort",
        "Complete paralysis"
    ],
    "notes": [
        "Non-ambulatory cohort",
        "Non-ambulatory cohort",
        "Non-ambulatory cohort",
        "Non-ambulatory cohort",
        "Non-ambulatory cohort",
        "Non-ambulatory >48 h cohort",
        "Tumor-specific HCC cohort",
        "Anterior vertebrectomy; selected thoracic metastases",
        "Tumor-specific NSCLC cohort",
        "Complete motor paralysis cohort"
    ]
})

data["p"] = data["event"] / data["n"]

clean6 = data[data["analysis_group"] == "Clean cohort"].copy()
expanded10 = data.copy()
paralysis = data[data["analysis_group"] == "Complete paralysis"].copy()

# ============================================================
# Meta-analysis function
# Random-effects model using logit transformation
# Use this only when k >= 2
# ============================================================

def run_meta(df, title, filename_prefix, x_label="Postoperative ambulatory recovery rate"):
    df = df.copy()
    df["p"] = df["event"] / df["n"]

    # Continuity correction
    df["event_cc"] = df["event"] + 0.5
    df["n_cc"] = df["n"] + 1
    df["nonevent_cc"] = df["n_cc"] - df["event_cc"]

    # Logit transformation
    df["p_cc"] = df["event_cc"] / df["n_cc"]
    df["logit"] = np.log(df["p_cc"] / (1 - df["p_cc"]))
    df["var"] = 1 / df["event_cc"] + 1 / df["nonevent_cc"]

    # Fixed-effect estimate
    w_fixed = 1 / df["var"]
    mean_fixed = np.sum(w_fixed * df["logit"]) / np.sum(w_fixed)

    # Heterogeneity
    Q = np.sum(w_fixed * (df["logit"] - mean_fixed) ** 2)
    dof = len(df) - 1

    tau2 = max(
        0,
        (Q - dof) / (np.sum(w_fixed) - np.sum(w_fixed ** 2) / np.sum(w_fixed))
    )

    # Random-effects estimate
    w_random = 1 / (df["var"] + tau2)
    mean_random = np.sum(w_random * df["logit"]) / np.sum(w_random)
    se_random = np.sqrt(1 / np.sum(w_random))

    def inv_logit(x):
        return np.exp(x) / (1 + np.exp(x))

    pooled = inv_logit(mean_random)
    ci_low = inv_logit(mean_random - 1.96 * se_random)
    ci_high = inv_logit(mean_random + 1.96 * se_random)

    I2 = max(0, (Q - dof) / Q) * 100 if Q > 0 else 0

    # Study-level CI for plotting
    df["se_p"] = np.sqrt((df["p"] * (1 - df["p"])) / df["n"])
    df["ci_low"] = np.maximum(0, df["p"] - 1.96 * df["se_p"])
    df["ci_high"] = np.minimum(1, df["p"] + 1.96 * df["se_p"])

    print("\n================================================")
    print(title)
    print("================================================")
    display(df[["Study", "n", "event", "p", "ci_low", "ci_high", "analysis_group", "notes"]])
    print(f"Pooled ambulatory recovery rate = {pooled:.3f}")
    print(f"95% CI = {ci_low:.3f}–{ci_high:.3f}")
    print(f"I² = {I2:.1f}%")
    print(f"Tau² = {tau2:.4f}")

    df[["Study", "n", "event", "p", "ci_low", "ci_high", "analysis_group", "notes"]].to_csv(
        f"{filename_prefix}_data.csv", index=False
    )

    # Forest plot
    plot_df = df.iloc[::-1].reset_index(drop=True)
    y = np.arange(len(plot_df))

    xerr_lower = plot_df["p"] - plot_df["ci_low"]
    xerr_upper = plot_df["ci_high"] - plot_df["p"]

    fig, ax = plt.subplots(figsize=FIGSIZE)

    ax.errorbar(
        plot_df["p"],
        y,
        xerr=[xerr_lower, xerr_upper],
        fmt="o",
        color="black",
        ecolor="black",
        elinewidth=1.5,
        capsize=5,
        markersize=6
    )

    ax.axvline(
        pooled,
        linestyle="--",
        color="black",
        linewidth=1.5,
        label=f"Pooled = {pooled:.2f}"
    )

    ax.set_yticks(y)
    ax.set_yticklabels(plot_df["Study"])
    ax.set_xlabel(x_label)
    ax.set_title(title)
    ax.set_xlim(0, 1.0)
    ax.legend(frameon=False)

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    plt.tight_layout()

    fig.savefig(f"{filename_prefix}.png", dpi=600, bbox_inches="tight")
    fig.savefig(f"{filename_prefix}.pdf", bbox_inches="tight")

    plt.show()

    return {
        "analysis": title,
        "number_of_studies": len(df),
        "total_patients": int(df["n"].sum()),
        "total_events": int(df["event"].sum()),
        "crude_recovery_rate": df["event"].sum() / df["n"].sum(),
        "pooled_ambulatory_recovery_rate": pooled,
        "ci_low": ci_low,
        "ci_high": ci_high,
        "I2": I2,
        "tau2": tau2,
        "Q": Q,
        "dof": dof,
        "analysis_type": "random-effects"
    }

# ============================================================
# Descriptive single-study function
# Use this for complete paralysis subgroup
# ============================================================

def run_descriptive_single_study(df, title, filename_prefix, x_label="Postoperative ambulatory recovery rate"):
    df = df.copy()
    df["p"] = df["event"] / df["n"]

    observed = df["event"].sum() / df["n"].sum()

    # Approximate binomial CI for display only
    se = np.sqrt(observed * (1 - observed) / df["n"].sum())
    ci_low = max(0, observed - 1.96 * se)
    ci_high = min(1, observed + 1.96 * se)

    df["ci_low"] = ci_low
    df["ci_high"] = ci_high

    print("\n================================================")
    print(title)
    print("================================================")
    display(df[["Study", "n", "event", "p", "ci_low", "ci_high", "analysis_group", "notes"]])
    print(f"Observed recovery rate = {observed:.3f}")
    print(f"95% CI = {ci_low:.3f}–{ci_high:.3f}")
    print("Note: This is a descriptive analysis because only one study was available.")

    df[["Study", "n", "event", "p", "ci_low", "ci_high", "analysis_group", "notes"]].to_csv(
        f"{filename_prefix}_data.csv", index=False
    )

    fig, ax = plt.subplots(figsize=FIGSIZE)

    y = np.arange(len(df))

    ax.errorbar(
        df["p"],
        y,
        xerr=[[df["p"].iloc[0] - ci_low], [ci_high - df["p"].iloc[0]]],
        fmt="o",
        color="black",
        ecolor="black",
        elinewidth=1.5,
        capsize=5,
        markersize=6
    )

    ax.axvline(
        observed,
        linestyle="--",
        color="black",
        linewidth=1.5,
        label=f"Observed recovery rate = {observed:.2f}"
    )

    ax.set_yticks(y)
    ax.set_yticklabels(df["Study"])
    ax.set_xlabel(x_label)
    ax.set_title(title)
    ax.set_xlim(0, 1.0)
    ax.legend(frameon=False)

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    plt.tight_layout()

    fig.savefig(f"{filename_prefix}.png", dpi=600, bbox_inches="tight")
    fig.savefig(f"{filename_prefix}.pdf", bbox_inches="tight")

    plt.show()

    return {
        "analysis": title,
        "number_of_studies": len(df),
        "total_patients": int(df["n"].sum()),
        "total_events": int(df["event"].sum()),
        "crude_recovery_rate": observed,
        "pooled_ambulatory_recovery_rate": observed,
        "ci_low": ci_low,
        "ci_high": ci_high,
        "I2": np.nan,
        "tau2": np.nan,
        "Q": np.nan,
        "dof": 0,
        "analysis_type": "descriptive"
    }

# ============================================================
# 1. Main analysis: clean 6 studies
# ============================================================

clean_result = run_meta(
    clean6,
    "Clean cohort analysis",
    "Figure1_clean6_forest_plot"
)

# ============================================================
# 2. Expanded analysis: 10 studies
# ============================================================

expanded_result = run_meta(
    expanded10,
    "Expanded analysis including additional abstract-based studies",
    "Figure2_expanded10_forest_plot"
)

# ============================================================
# 3. Complete paralysis subgroup
# Descriptive only because k = 1
# ============================================================

paralysis_result = run_descriptive_single_study(
    paralysis,
    "Complete paralysis subgroup (descriptive analysis)",
    "Figure3_complete_paralysis_forest_plot"
)

# ============================================================
# 4. Sensitivity analysis: Expanded analysis excluding Gokaslan
# ============================================================

expanded_no_gokaslan = expanded10[expanded10["Study"] != "Gokaslan 1998"].copy()

expanded_no_gokaslan_result = run_meta(
    expanded_no_gokaslan,
    "Sensitivity analysis: expanded cohort excluding Gokaslan",
    "Figure4_expanded_excluding_Gokaslan"
)

# ============================================================
# 5. Summary figure
# ============================================================

summary_results = pd.DataFrame([
    clean_result,
    expanded_result,
    paralysis_result,
    expanded_no_gokaslan_result
])

summary_results["rate_percent"] = summary_results["pooled_ambulatory_recovery_rate"] * 100
summary_results["ci_low_percent"] = summary_results["ci_low"] * 100
summary_results["ci_high_percent"] = summary_results["ci_high"] * 100

label_map = {
    "Clean cohort analysis": "Clean cohort\nanalysis",
    "Expanded analysis including additional abstract-based studies": "Expanded analysis",
    "Complete paralysis subgroup (descriptive analysis)": "Complete paralysis\nsubgroup",
    "Sensitivity analysis: expanded cohort excluding Gokaslan": "Sensitivity analysis\n(excluding Gokaslan)"
}

summary_results["display_label"] = summary_results["analysis"].map(label_map)

order = [
    "Clean cohort analysis",
    "Expanded analysis including additional abstract-based studies",
    "Complete paralysis subgroup (descriptive analysis)",
    "Sensitivity analysis: expanded cohort excluding Gokaslan"
]

plot_summary = summary_results.set_index("analysis").loc[order].reset_index()
plot_summary = plot_summary.iloc[::-1].reset_index(drop=True)

y = np.arange(len(plot_summary))

xerr_lower = plot_summary["rate_percent"] - plot_summary["ci_low_percent"]
xerr_upper = plot_summary["ci_high_percent"] - plot_summary["rate_percent"]

fig, ax = plt.subplots(figsize=(9, 6))

ax.errorbar(
    plot_summary["rate_percent"],
    y,
    xerr=[xerr_lower, xerr_upper],
    fmt="o",
    color="black",
    ecolor="black",
    elinewidth=1.8,
    capsize=6,
    markersize=8
)

ax.set_yticks(y)
ax.set_yticklabels(plot_summary["display_label"], fontsize=14)
ax.set_xlabel("Ambulatory recovery rate (%)", fontsize=15)
ax.set_title("Summary of meta-analytic estimates", fontsize=18)
ax.set_xlim(0, 100)
ax.set_xticks([0, 25, 50, 75, 100])

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.subplots_adjust(left=0.30, right=0.96, top=0.88, bottom=0.16)

fig.savefig("Figure5_summary_clean_expanded_paralysis.png", dpi=600, bbox_inches="tight")
fig.savefig("Figure5_summary_clean_expanded_paralysis.pdf", bbox_inches="tight")

plt.show()

# ============================================================
# 6. Leave-one-out sensitivity analysis for expanded 10
# ============================================================

loo_results = []

for study in expanded10["Study"]:
    temp = expanded10[expanded10["Study"] != study].copy()
    result = run_meta(
        temp,
        f"Leave-one-out analysis: excluding {study}",
        f"LOO_expanded_excluding_{study.replace(' ', '_')}"
    )
    result["excluded_study"] = study
    loo_results.append(result)

loo_summary = pd.DataFrame(loo_results)
loo_summary.to_csv("Table_leave_one_out_expanded10.csv", index=False)

print("\n===== Leave-one-out sensitivity summary: expanded 10 =====")
display(loo_summary)

# ============================================================
# 7. Funnel plot for expanded 10
# ============================================================

funnel = expanded10.copy()
funnel["p"] = funnel["event"] / funnel["n"]
funnel["se_p"] = np.sqrt((funnel["p"] * (1 - funnel["p"])) / funnel["n"])
funnel["precision"] = 1 / funnel["se_p"]

fig, ax = plt.subplots(figsize=FIGSIZE)

ax.scatter(
    funnel["p"],
    funnel["precision"],
    color="black",
    s=55
)

ax.set_xlabel("Ambulatory recovery rate")
ax.set_ylabel("Precision")
ax.set_title("Funnel plot: expanded analysis")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()

fig.savefig("Figure6_funnel_plot_expanded10.png", dpi=600, bbox_inches="tight")
fig.savefig("Figure6_funnel_plot_expanded10.pdf", bbox_inches="tight")

plt.show()

# ============================================================
# 8. Egger's regression test
# Caution: limited interpretability with small number of studies
# ============================================================

funnel["SND"] = funnel["p"] / funnel["se_p"]
funnel["precision_for_egger"] = 1 / funnel["se_p"]

slope, intercept, r, p_value, std_err = stats.linregress(
    funnel["precision_for_egger"],
    funnel["SND"]
)

print("\n===== Publication bias =====")
print(f"Egger's regression intercept = {intercept:.3f}")
print(f"Egger's test p-value = {p_value:.3f}")
print("Note: Egger's test is exploratory because the number of studies is small.")

with open("Egger_test_expanded10.txt", "w") as f:
    f.write(f"Egger's regression intercept = {intercept:.3f}\n")
    f.write(f"Egger's test p-value = {p_value:.3f}\n")
    f.write("Note: Egger's test is exploratory because the number of studies is small.\n")

# ============================================================
# 9. Manuscript Table 1: Study characteristics
# ============================================================

table1 = data.copy()
table1["Recovery rate"] = (table1["event"] / table1["n"] * 100).map(lambda x: f"{x:.1f}%")

table1 = table1.rename(columns={
    "n": "Sample size",
    "event": "Ambulatory recovery events",
    "analysis_group": "Analysis category",
    "notes": "Cohort notes"
})

table1 = table1[[
    "Study",
    "Sample size",
    "Ambulatory recovery events",
    "Recovery rate",
    "Analysis category",
    "Cohort notes"
]]

table1.to_csv("Table1_study_characteristics_10studies.csv", index=False)

print("\n===== Table 1. Study characteristics =====")
display(table1)

# ============================================================
# 10. Manuscript Table 2: Meta-analysis results
# ============================================================

table2 = summary_results.copy()

def format_ci(row):
    if row["analysis_type"] == "descriptive":
        return "—"
    return f"{row['ci_low']*100:.1f}–{row['ci_high']*100:.1f}%"

def format_i2(row):
    if row["analysis_type"] == "descriptive":
        return "—"
    return f"{row['I2']:.1f}%"

def format_tau2(row):
    if row["analysis_type"] == "descriptive":
        return "—"
    return f"{row['tau2']:.4f}"

table2_for_manuscript = pd.DataFrame({
    "Analysis": table2["analysis"],
    "Number of studies": table2["number_of_studies"],
    "Total patients": table2["total_patients"],
    "Total recovery events": table2["total_events"],
    "Crude recovery rate": table2["crude_recovery_rate"].map(lambda x: f"{x*100:.1f}%"),
    "Pooled/observed ambulatory recovery rate": table2["pooled_ambulatory_recovery_rate"].map(lambda x: f"{x*100:.1f}%"),
    "95% CI": table2.apply(format_ci, axis=1),
    "I²": table2.apply(format_i2, axis=1),
    "Tau²": table2.apply(format_tau2, axis=1)
})

table2_for_manuscript.to_csv("Table2_meta_analysis_results_clean_expanded_paralysis.csv", index=False)

print("\n===== Table 2. Meta-analysis results =====")
display(table2_for_manuscript)

# ============================================================
# 11. Manuscript Table 3: Leave-one-out sensitivity
# ============================================================

loo_table = loo_summary.copy()

loo_table_for_manuscript = pd.DataFrame({
    "Excluded study": loo_table["excluded_study"],
    "Number of studies": loo_table["number_of_studies"],
    "Total patients": loo_table["total_patients"],
    "Pooled ambulatory recovery rate": loo_table["pooled_ambulatory_recovery_rate"].map(lambda x: f"{x*100:.1f}%"),
    "95% CI": [
        f"{lo*100:.1f}–{hi*100:.1f}%"
        for lo, hi in zip(loo_table["ci_low"], loo_table["ci_high"])
    ],
    "I²": loo_table["I2"].map(lambda x: f"{x:.1f}%")
})

loo_table_for_manuscript.to_csv("Table3_leave_one_out_sensitivity_expanded10.csv", index=False)

print("\n===== Table 3. Leave-one-out sensitivity analysis =====")
display(loo_table_for_manuscript)

print("\nAll figures and manuscript tables have been saved.")
```
