```python
# ============================================================
# Meta-analysis: Ambulatory recovery after surgery for MESCC
# Clean cohort: 6 studies
# Google Colab version
# Black-and-white / unified fonts / same figure size
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# ============================================================
# Global style
# ============================================================

BASE_FONT = 14
TITLE_FONT = 16
FIGSIZE = (8, 8)   # Do not change

plt.rcParams.update({
    "font.size": BASE_FONT,
    "axes.titlesize": TITLE_FONT,
    "axes.labelsize": BASE_FONT,
    "xtick.labelsize": BASE_FONT,
    "ytick.labelsize": BASE_FONT,
    "legend.fontsize": BASE_FONT,
    "font.family": "DejaVu Sans",
    "axes.edgecolor": "black",
    "axes.linewidth": 1.2,
    "xtick.color": "black",
    "ytick.color": "black",
    "text.color": "black",
    "axes.labelcolor": "black",
})

# ============================================================
# Dataset
# event = postoperative ambulatory recovery
# ============================================================

data = pd.DataFrame({
    "Study": [
        "Kim 2011",
        "Tateiwa 2018",
        "Ohashi 2017",
        "Kinoshita 2023",
        "Li 2024",
        "Wänman 2025"
    ],
    "n": [57, 31, 82, 113, 37, 111],
    "event": [39, 17, 50, 70, 22, 61]
})

data["p"] = data["event"] / data["n"]

# ============================================================
# Meta-analysis function
# Random-effects model using logit transformation
# ============================================================

def run_meta(df, title, filename_prefix):
    df = df.copy()
    df["p"] = df["event"] / df["n"]

    # Continuity correction
    df["event_cc"] = df["event"] + 0.5
    df["n_cc"] = df["n"] + 1
    df["nonevent_cc"] = df["n_cc"] - df["event_cc"]

    # Logit transformation
    df["p_cc"] = df["event_cc"] / df["n_cc"]
    df["logit"] = np.log(df["p_cc"] / (1 - df["p_cc"]))
    df["var"] = 1 / df["event_cc"] + 1 / df["nonevent_cc"]

    # Fixed-effect estimate
    w_fixed = 1 / df["var"]
    mean_fixed = np.sum(w_fixed * df["logit"]) / np.sum(w_fixed)

    # Heterogeneity
    Q = np.sum(w_fixed * (df["logit"] - mean_fixed) ** 2)
    dof = len(df) - 1

    tau2 = max(
        0,
        (Q - dof) / (np.sum(w_fixed) - np.sum(w_fixed ** 2) / np.sum(w_fixed))
    )

    # Random-effects estimate
    w_random = 1 / (df["var"] + tau2)
    mean_random = np.sum(w_random * df["logit"]) / np.sum(w_random)
    se_random = np.sqrt(1 / np.sum(w_random))

    def inv_logit(x):
        return np.exp(x) / (1 + np.exp(x))

    pooled = inv_logit(mean_random)
    ci_low = inv_logit(mean_random - 1.96 * se_random)
    ci_high = inv_logit(mean_random + 1.96 * se_random)

    I2 = max(0, (Q - dof) / Q) * 100 if Q > 0 else 0

    # Study-level normal CI for plotting
    df["se_p"] = np.sqrt((df["p"] * (1 - df["p"])) / df["n"])
    df["ci_low"] = np.maximum(0, df["p"] - 1.96 * df["se_p"])
    df["ci_high"] = np.minimum(1, df["p"] + 1.96 * df["se_p"])

    print("\n================================================")
    print(title)
    print("================================================")
    display(df[["Study", "n", "event", "p", "ci_low", "ci_high"]])
    print(f"Pooled ambulatory recovery rate = {pooled:.3f}")
    print(f"95% CI = {ci_low:.3f}–{ci_high:.3f}")
    print(f"I² = {I2:.1f}%")
    print(f"Tau² = {tau2:.4f}")

    df[["Study", "n", "event", "p", "ci_low", "ci_high"]].to_csv(
        f"{filename_prefix}_data.csv", index=False
    )

    # Forest plot
    plot_df = df.iloc[::-1].reset_index(drop=True)
    y = np.arange(len(plot_df))

    xerr_lower = plot_df["p"] - plot_df["ci_low"]
    xerr_upper = plot_df["ci_high"] - plot_df["p"]

    fig, ax = plt.subplots(figsize=FIGSIZE)

    ax.errorbar(
        plot_df["p"],
        y,
        xerr=[xerr_lower, xerr_upper],
        fmt="o",
        color="black",
        ecolor="black",
        elinewidth=1.5,
        capsize=5,
        markersize=6
    )

    ax.axvline(
        pooled,
        linestyle="--",
        color="black",
        linewidth=1.5,
        label=f"Pooled = {pooled:.2f}"
    )

    ax.set_yticks(y)
    ax.set_yticklabels(plot_df["Study"])
    ax.set_xlabel("Postoperative ambulatory recovery rate")
    ax.set_title(title)
    ax.set_xlim(0, 1.0)
    ax.legend(frameon=False)

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    plt.tight_layout()

    fig.savefig(f"{filename_prefix}.png", dpi=600, bbox_inches="tight")
    fig.savefig(f"{filename_prefix}.pdf", bbox_inches="tight")

    plt.show()

    return {
        "analysis": title,
        "number_of_studies": len(df),
        "total_patients": int(df["n"].sum()),
        "total_events": int(df["event"].sum()),
        "pooled_ambulatory_recovery_rate": pooled,
        "ci_low": ci_low,
        "ci_high": ci_high,
        "I2": I2,
        "tau2": tau2,
        "Q": Q,
        "dof": dof
    }

# ============================================================
# Main analysis: clean 6 studies
# ============================================================

main_result = run_meta(
    data,
    "Ambulatory recovery after surgery for MESCC",
    "Figure_clean6_forest_plot"
)

# ============================================================
# Leave-one-out sensitivity analysis
# ============================================================

loo_results = []

for study in data["Study"]:
    temp = data[data["Study"] != study].copy()
    result = run_meta(
        temp,
        f"Leave-one-out analysis: excluding {study}",
        f"LOO_excluding_{study.replace(' ', '_').replace('ä', 'a')}"
    )
    result["excluded_study"] = study
    loo_results.append(result)

loo_summary = pd.DataFrame(loo_results)
loo_summary.to_csv("Table_leave_one_out_sensitivity.csv", index=False)

print("\n===== Leave-one-out sensitivity summary =====")
display(loo_summary)

# ============================================================
# Summary table
# ============================================================

summary = pd.DataFrame([main_result])
summary.to_csv("Table_meta_analysis_summary_clean6.csv", index=False)

print("\n===== Meta-analysis summary =====")
display(summary)

# ============================================================
# Funnel plot
# ============================================================

data["se_p"] = np.sqrt((data["p"] * (1 - data["p"])) / data["n"])
data["precision"] = 1 / data["se_p"]

fig, ax = plt.subplots(figsize=FIGSIZE)

ax.scatter(
    data["p"],
    data["precision"],
    color="black",
    s=55
)

ax.set_xlabel("Ambulatory recovery rate")
ax.set_ylabel("Precision")
ax.set_title("Funnel plot")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()

fig.savefig("Figure_funnel_plot_clean6.png", dpi=600, bbox_inches="tight")
fig.savefig("Figure_funnel_plot_clean6.pdf", bbox_inches="tight")

plt.show()

# ============================================================
# Manuscript tables
# ============================================================

table1 = pd.DataFrame({
    "Study": [
        "Kim 2011",
        "Tateiwa 2018",
        "Ohashi 2017",
        "Kinoshita 2023",
        "Li 2024",
        "Wänman 2025"
    ],
    "Sample size": [57, 31, 82, 113, 37, 111],
    "Ambulatory recovery events": [39, 17, 50, 70, 22, 61],
    "Recovery rate": [
        "68.4%",
        "54.8%",
        "61.0%",
        "61.9%",
        "59.5%",
        "55.0%"
    ],
    "Population type": [
        "Non-ambulatory cohort",
        "Non-ambulatory cohort",
        "Non-ambulatory cohort",
        "Non-ambulatory subgroup within neurological impairment cohort",
        "Non-ambulatory cohort",
        "Non-ambulatory cohort >48 h"
    ],
    "Ambulation definition": [
        "Nurick-based independent ambulation",
        "Frankel-based ambulatory status",
        "Walking >10 m with or without aid",
        "Walking with walker/cane/crutches included",
        "Frankel D/E or postoperative walking ability",
        "Frankel D/E at 1 month"
    ]
})

table1.to_csv("Table1_clean6_study_characteristics.csv", index=False)

table2 = pd.DataFrame({
    "Analysis": ["Clean cohort analysis"],
    "Number of studies": [main_result["number_of_studies"]],
    "Total patients": [main_result["total_patients"]],
    "Total recovery events": [main_result["total_events"]],
    "Pooled ambulatory recovery rate": [
        f"{main_result['pooled_ambulatory_recovery_rate']*100:.1f}%"
    ],
    "95% CI": [
        f"{main_result['ci_low']*100:.1f}–{main_result['ci_high']*100:.1f}%"
    ],
    "I²": [f"{main_result['I2']:.1f}%"],
    "Tau²": [f"{main_result['tau2']:.4f}"]
})

table2.to_csv("Table2_clean6_meta_analysis_results.csv", index=False)

print("\n===== Table 1. Study characteristics =====")
display(table1)

print("\n===== Table 2. Meta-analysis results =====")
display(table2)

print("\nAll figures and tables have been saved.")
```
