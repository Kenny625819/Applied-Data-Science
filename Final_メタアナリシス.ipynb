```python
# ============================================================
# FINAL SUBMISSION CODE
# Meta-analysis of postoperative ambulatory recovery
# Black-and-white / unified fonts
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# ============================================================
# Global style
# ============================================================

BASE_FONT = 14
TITLE_FONT = 16
FIGSIZE = (8, 8)

plt.rcParams.update({
    "font.size": BASE_FONT,
    "axes.titlesize": TITLE_FONT,
    "axes.labelsize": BASE_FONT,
    "xtick.labelsize": BASE_FONT,
    "ytick.labelsize": BASE_FONT,
    "legend.fontsize": BASE_FONT,
    "font.family": "DejaVu Sans",
    "axes.edgecolor": "black",
    "axes.linewidth": 1.2,
    "xtick.color": "black",
    "ytick.color": "black",
    "text.color": "black",
    "axes.labelcolor": "black",
})

# ============================================================
# Dataset
# event = postoperative ambulatory recovery
# n = preoperative non-ambulatory patients
# ============================================================

data = pd.DataFrame({
    "Study": [
        "Watanabe 2016",
        "Factors 2022",
        "Single-stage decompression 2012",
        "Timing study 2016",
        "Li 2024",
        "Non-fusion palliative surgery 2021",
        "Resumption study 2011",
        "Effect of timing 2013"
    ],
    "Design": [
        "Retrospective",
        "Retrospective",
        "Retrospective",
        "Retrospective",
        "Retrospective",
        "Retrospective",
        "Retrospective",
        "Retrospective"
    ],
    "Population": [
        "MESCC",
        "MESCC",
        "MESCC",
        "MESCC",
        "MESCC",
        "MSCC",
        "Spinal metastasis",
        "MSCC"
    ],
    "Baseline": [
        "Non-ambulatory",
        "Non-ambulatory",
        "Non-ambulatory",
        "Non-ambulatory",
        "Non-ambulatory",
        "Non-ambulatory",
        "Non-ambulatory",
        "Non-ambulatory"
    ],
    "Outcome": [
        "Ambulatory recovery",
        "Ambulatory recovery",
        "Ambulatory recovery",
        "Ambulatory recovery",
        "Ambulatory recovery",
        "Ambulatory recovery",
        "Ambulatory recovery",
        "Ambulatory recovery"
    ],
    "n": [10, 86, 21, 34, 37, 19, 57, 9],
    "event": [2, 21, 10, 18, 22, 12, 39, 7]
})

data["rate"] = data["event"] / data["n"]

# ============================================================
# Meta-analysis function
# Random-effects model using logit transformation
# ============================================================

def run_meta(df, title, filename_prefix):
    df = df.copy()
    df["rate"] = df["event"] / df["n"]

    # Continuity correction
    df["event_cc"] = df["event"] + 0.5
    df["n_cc"] = df["n"] + 1
    df["nonevent_cc"] = df["n_cc"] - df["event_cc"]

    # Logit transformation
    df["p_cc"] = df["event_cc"] / df["n_cc"]
    df["logit"] = np.log(df["p_cc"] / (1 - df["p_cc"]))
    df["var"] = 1 / df["event_cc"] + 1 / df["nonevent_cc"]

    # Fixed-effect estimate
    w_fixed = 1 / df["var"]
    mean_fixed = np.sum(w_fixed * df["logit"]) / np.sum(w_fixed)

    # Heterogeneity
    Q = np.sum(w_fixed * (df["logit"] - mean_fixed) ** 2)
    dof = len(df) - 1

    tau2 = max(
        0,
        (Q - dof) / (np.sum(w_fixed) - np.sum(w_fixed ** 2) / np.sum(w_fixed))
    )

    # Random-effects estimate
    w_random = 1 / (df["var"] + tau2)
    mean_random = np.sum(w_random * df["logit"]) / np.sum(w_random)
    se_random = np.sqrt(1 / np.sum(w_random))

    def inv_logit(x):
        return np.exp(x) / (1 + np.exp(x))

    pooled = inv_logit(mean_random)
    ci_low = inv_logit(mean_random - 1.96 * se_random)
    ci_high = inv_logit(mean_random + 1.96 * se_random)

    I2 = max(0, (Q - dof) / Q) * 100 if Q > 0 else 0

    # Study-level normal CI for plotting
    df["se_rate"] = np.sqrt((df["rate"] * (1 - df["rate"])) / df["n"])
    df["ci_low"] = np.maximum(0, df["rate"] - 1.96 * df["se_rate"])
    df["ci_high"] = np.minimum(1, df["rate"] + 1.96 * df["se_rate"])

    # Print detailed results
    print("\n================================================")
    print(title)
    print("================================================")
    display(df[["Study", "Design", "Population", "Baseline", "Outcome", "n", "event", "rate", "ci_low", "ci_high"]])
    print(f"Pooled ambulatory recovery rate = {pooled:.3f}")
    print(f"95% CI = {ci_low:.3f}–{ci_high:.3f}")
    print(f"I² = {I2:.1f}%")
    print(f"Tau² = {tau2:.4f}")

    # Save study-level result table
    df[["Study", "Design", "Population", "Baseline", "Outcome", "n", "event", "rate", "ci_low", "ci_high"]].to_csv(
        f"{filename_prefix}_data.csv", index=False
    )

    # Forest plot
    plot_df = df.iloc[::-1].reset_index(drop=True)
    y = np.arange(len(plot_df))

    xerr_lower = plot_df["rate"] - plot_df["ci_low"]
    xerr_upper = plot_df["ci_high"] - plot_df["rate"]

    fig, ax = plt.subplots(figsize=FIGSIZE)

    ax.errorbar(
        plot_df["rate"],
        y,
        xerr=[xerr_lower, xerr_upper],
        fmt="o",
        color="black",
        ecolor="black",
        elinewidth=1.5,
        capsize=5,
        markersize=6
    )

    ax.axvline(
        pooled,
        linestyle="--",
        color="black",
        linewidth=1.5,
        label=f"Pooled = {pooled:.2f}"
    )

    ax.set_yticks(y)
    ax.set_yticklabels(plot_df["Study"])
    ax.set_xlabel("Postoperative ambulatory recovery rate")
    ax.set_title(title)
    ax.set_xlim(0, 1.0)
    ax.legend(frameon=False)

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    plt.tight_layout()

    fig.savefig(f"{filename_prefix}.png", dpi=600, bbox_inches="tight")
    fig.savefig(f"{filename_prefix}.pdf", bbox_inches="tight")

    plt.show()

    return {
        "analysis": title,
        "number_of_studies": len(df),
        "total_patients": int(df["n"].sum()),
        "total_events": int(df["event"].sum()),
        "pooled_ambulatory_recovery_rate": pooled,
        "ci_low": ci_low,
        "ci_high": ci_high,
        "I2": I2,
        "tau2": tau2,
        "Q": Q,
        "dof": dof
    }

# ============================================================
# Main analysis
# ============================================================

main_result = run_meta(
    data,
    "Postoperative ambulatory recovery after surgery",
    "Figure1_ambulatory_recovery_forest_plot"
)

# ============================================================
# Sensitivity analysis
# Excluding very small studies: n < 10
# ============================================================

sensitivity_data = data[data["n"] >= 10].copy()

sensitivity_result = run_meta(
    sensitivity_data,
    "Sensitivity analysis: excluding studies with n < 10",
    "Figure2_sensitivity_excluding_small_studies"
)

# ============================================================
# Summary table
# ============================================================

summary = pd.DataFrame([
    main_result,
    sensitivity_result
])

summary.to_csv("Table_meta_analysis_summary.csv", index=False)

print("\n===== Meta-analysis summary =====")
display(summary)

# ============================================================
# Funnel plot
# ============================================================

funnel_data = data.copy()
funnel_data["rate"] = funnel_data["event"] / funnel_data["n"]
funnel_data["se_rate"] = np.sqrt((funnel_data["rate"] * (1 - funnel_data["rate"])) / funnel_data["n"])

fig, ax = plt.subplots(figsize=FIGSIZE)

ax.scatter(
    funnel_data["rate"],
    funnel_data["se_rate"],
    color="black",
    s=55
)

ax.set_xlabel("Ambulatory recovery rate")
ax.set_ylabel("Standard error")
ax.set_title("Funnel plot")
ax.invert_yaxis()

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()

fig.savefig("Figure3_funnel_plot.png", dpi=600, bbox_inches="tight")
fig.savefig("Figure3_funnel_plot.pdf", bbox_inches="tight")

plt.show()

# ============================================================
# Egger's regression test
# Note: interpret cautiously because only 8 studies are included
# ============================================================

egger_data = data.copy()
egger_data["rate"] = egger_data["event"] / egger_data["n"]
egger_data["se"] = np.sqrt((egger_data["rate"] * (1 - egger_data["rate"])) / egger_data["n"])
egger_data["precision"] = 1 / egger_data["se"]
egger_data["standardized_effect"] = egger_data["rate"] / egger_data["se"]

slope, intercept, r, p_value, std_err = stats.linregress(
    egger_data["precision"],
    egger_data["standardized_effect"]
)

print("\n===== Publication bias =====")
print(f"Egger's test p-value = {p_value:.3f}")
print("Note: Egger's test should be interpreted cautiously because fewer than 10 studies were included.")

with open("Egger_test_result.txt", "w") as f:
    f.write(f"Egger's test p-value = {p_value:.3f}\n")
    f.write("Note: Egger's test should be interpreted cautiously because fewer than 10 studies were included.\n")

# ============================================================
# Manuscript tables
# ============================================================

table1 = data.copy()
table1["Rate"] = table1["rate"].apply(lambda x: f"{x*100:.1f}%")
table1 = table1[[
    "Study",
    "Design",
    "Population",
    "Baseline",
    "Outcome",
    "n",
    "event",
    "Rate"
]]

table1.columns = [
    "Study",
    "Study design",
    "Population",
    "Baseline",
    "Outcome",
    "n",
    "Event",
    "Rate"
]

table1.to_csv("Table1_study_characteristics.csv", index=False)

table2 = pd.DataFrame({
    "Analysis": [
        "Main analysis",
        "Sensitivity analysis excluding studies with n < 10"
    ],
    "Studies": [
        main_result["number_of_studies"],
        sensitivity_result["number_of_studies"]
    ],
    "Patients": [
        main_result["total_patients"],
        sensitivity_result["total_patients"]
    ],
    "Events": [
        main_result["total_events"],
        sensitivity_result["total_events"]
    ],
    "Pooled ambulatory recovery rate": [
        f"{main_result['pooled_ambulatory_recovery_rate']*100:.1f}%",
        f"{sensitivity_result['pooled_ambulatory_recovery_rate']*100:.1f}%"
    ],
    "95% CI": [
        f"{main_result['ci_low']*100:.1f}–{main_result['ci_high']*100:.1f}%",
        f"{sensitivity_result['ci_low']*100:.1f}–{sensitivity_result['ci_high']*100:.1f}%"
    ],
    "I²": [
        f"{main_result['I2']:.1f}%",
        f"{sensitivity_result['I2']:.1f}%"
    ],
    "Tau²": [
        f"{main_result['tau2']:.4f}",
        f"{sensitivity_result['tau2']:.4f}"
    ]
})

table2.to_csv("Table2_meta_analysis_results.csv", index=False)

print("\n===== Table 1. Study characteristics =====")
display(table1)

print("\n===== Table 2. Meta-analysis results =====")
display(table2)

# ============================================================
# Save all extracted data as Excel
# ============================================================

with pd.ExcelWriter("Meta_analysis_outputs.xlsx") as writer:
    table1.to_excel(writer, sheet_name="Study characteristics", index=False)
    table2.to_excel(writer, sheet_name="Meta-analysis results", index=False)
    summary.to_excel(writer, sheet_name="Raw summary", index=False)
    data.to_excel(writer, sheet_name="Raw extracted data", index=False)

print("\nAll submission-ready figures and tables have been saved.")
print("Saved files:")
print("- Figure1_ambulatory_recovery_forest_plot.png / .pdf")
print("- Figure2_sensitivity_excluding_small_studies.png / .pdf")
print("- Figure3_funnel_plot.png / .pdf")
print("- Table1_study_characteristics.csv")
print("- Table2_meta_analysis_results.csv")
print("- Table_meta_analysis_summary.csv")
print("- Egger_test_result.txt")
print("- Meta_analysis_outputs.xlsx")
```
